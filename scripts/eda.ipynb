```python
# Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Configure visualizations
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (12, 6)
```

```python
# Load processed data
file_path = "D:\CLG\SEM-V\BDA\data\processed_food_data.csv"  # Adjust path if needed
df = pd.read_csv(file_path)

# Display basic information about the dataset
print("Dataset Information:")
print(df.info())
print("\nFirst few rows:")
print(df.head())
```

```python
# Statistical summary of the dataset
print("Statistical Summary:")
print(df.describe())

# Check for missing values
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# Line plot for historical prices over time by food item
df["Date"] = pd.to_datetime(df["Date"])  # Ensure Date column is in datetime format
df.sort_values("Date", inplace=True)

plt.figure(figsize=(14, 8))
sns.lineplot(data=df, x="Date", y="Historical_Price", hue="Food_Item")
plt.title("Historical Prices Over Time by Food Item")
plt.xlabel("Date")
plt.ylabel("Historical Price")
plt.xticks(rotation=45)
plt.legend(title="Food Item", loc="upper left")
plt.show()
```

```python
# Correlation heatmap for numerical features
plt.figure(figsize=(10, 8))
correlation = df.corr()
sns.heatmap(correlation, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Boxplot to check for outliers in Historical Price
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x="Food_Item", y="Historical_Price", palette="Set3")
plt.title("Price Distribution by Food Item")
plt.xlabel("Food Item")
plt.ylabel("Historical Price")
plt.show()
```

```python
# Scatter plot for Rainfall vs Historical Price
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="Rainfall", y="Historical_Price", hue="Food_Item", palette="viridis")
plt.title("Rainfall vs Historical Price")
plt.xlabel("Rainfall (mm)")
plt.ylabel("Historical Price")
plt.legend(title="Food Item", loc="upper right")
plt.show()
```

```python
# Distribution of Historical Prices
plt.figure(figsize=(10, 6))
sns.histplot(df["Historical_Price"], kde=True, bins=30, color="blue")
plt.title("Distribution of Historical Prices")
plt.xlabel("Historical Price")
plt.ylabel("Frequency")
plt.show()
```
